```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how="left")
```

```python
final_df.head(2)
```

```python
#another way to check the datatypes 
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values= final_df.Country.value_counts().values
```

```python
##pie chart
plt.pie(country_values,labels=country_names)
```

```python
#above looking very goosy so looking for only first three/ most three transaction
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%") # <- to see in what percentage 
```

```python
#observation:Zomato maximum records or transactions are from india ,usa ,uk
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating',"Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

```python
## observation 
# when ratings between 4.5 to 4.9 ----. Excellent
# when ratings between 4.0 to 3.4 =---->very good
# when ratings between 3.5 to 3.9 ---> good 
# when ratings between 3.0 to 3.4 ---> average
# when ratings between 2.5 to 2.9 -----. average
# when ratings between 2.0 to 2.4 -----. average
```

```python
ratings
```

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
# to increase the size of diagram we have a code in matplotlib
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
#to do color mapping 
sns.barplot(x="Aggregate rating",y="Rating Count",hue= "Rating color",data=ratings)
```

```python
# but  color name is not what it is so doing color mapping
sns.barplot(x="Aggregate rating",y="Rating Count",hue= "Rating color",data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```

```python
##observation: 1) Not rated count is very high
#2) max rating are between 2.5 to 3.4  
```

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Green'])
```
